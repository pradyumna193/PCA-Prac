```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",header = None, names = col)
col = ["SL", "SW", "PL", "PW", "Class"]
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
x = df.iloc[:,:-1]
y = df["Class"]
```

```python
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25, random_state = 25)
```

```python
lg = LogisticRegression()
lg.fit(xtrain ,ytrain)
ypred = lg.predict(xtest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, ypred))
```

```python
#PCA
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
sc = StandardScaler()

x = sc.fit_transform(x)
x
```

```python
pca = PCA()
pcs = pca.fit_transform(x)
```

```python
pcs
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_));
```

```python
pca = PCA(n_components = 2)
pcs = pca.fit_transform(x)
pcs
```

```python
data = pd.DataFrame(pcs, columns = ["PC1", "PC2"])
data
```

```python
data["Class"] = y
```

```python
data
```

```python
x = data.iloc[:, :2]
y = data["Class"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x , y, test_size = 0.25, random_state = 25)
```

```python
lg = LogisticRegression()

lg.fit(xtrain, ytrain)
ypred = lg.predict(xtest)

print(classification_report(ytest,ypred))
```
